## Câu 2:
Explain the significance of Term Frequency (TF) and Inverse Document Frequency (IDF) in text analysis. Why is it essential to use both TF and IDF together?
* Term Frequency (TF): Nó đo lường tần suất một từ xuất hiện trong một tài liệu. Nó được tính bằng cách đếm số lần một từ xuất hiện trong tài liệu và sau đó chia cho tổng số từ trong tài liệu đó. TF cho chúng ta một ý tưởng về sự quan trọng của một từ trong một tài liệu cụ thể.
* Inverse Document Frequency (IDF): IDF đo lường sự quan trọng của một từ trên toàn bộ tập hợp các tài liệu. Nó được tính bằng cách lấy log của tổng số tài liệu chia cho số tài liệu chứa từ đó. IDF giúp xác định các từ hiếm do đó cho biết 1 từ có thể mang nhiều thông tin hơn.
* Quan trọng phải sử dụng cả TF và IDF cùng nhau vì chúng bổ sung cho nhau. TF bắt kịp sự quan trọng của một từ trong một tài liệu cụ thể, trong khi IDF bắt kịp sự quan trọng của một từ trên toàn bộ tập hợp các tài liệu. Kết hợp TF và IDF trong công thức TF-IDF giúp ưu tiên những từ vừa thường xuyên xuất hiện trong một tài liệu và vừa hiếm trên toàn bộ tập hợp, làm cho nó trở thành một công cụ quý báu cho phân tích văn bản, truy xuất thông tin và xếp hạng tài liệu.

* Công thức IDF trong scikit-learn được triển khai như sau:

    IDF(t) = log((1 + n) / (1 + df(t))) + 1
    Trong đó:
    * t là một từ cụ thể trong bộ sưu tập tài liệu.
    * n là tổng số tài liệu trong bộ sưu tập tài liệu.
    * df(t) là số tài liệu chứa từ t.
* Công thức công thức IDF tiêu chuẩn:

    IDF(t) = log(N / df(t))
    Trong đó:
    * N là tổng số tài liệu trong bộ sưu tập tài liệu.
    * df(t) là số tài liệu chứa từ t.

## Câu 3
Write a program to tokenize the words in the reviews, lovercase all words, remove special characters and remove the stopwords.

```python
def lovercase_all_words(text):
    text = text.lower()
    return text
def remove_tag_sentences(text):
    pattern = r'<[^>]+>'
    text = re.sub(pattern, "",text)
    return text
def remove_special_characters(sentence):
    pattern = r'[^a-zA-Z0-9\s]'
    cleaned_sentence = re.sub(pattern, '', sentence)
    return cleaned_sentence
def remove_stopwords(sentence):
    stop_words = set(stopwords.words('english'))
    words = sentence.split()
    filtered_sentence = [word for word in words if word not in stop_words]
    return ' '.join(filtered_sentence)
def lematize_word(list_word):
    lemmatizer = WordNetLemmatizer()
    for i in range(len(list_word)):
        list_word[i] = lemmatizer.lemmatize(list_word[i])
    return list_word
def tokenze_sentences(sentences):
    sentences = lovercase_all_words(sentences)
    sentences = remove_tag_sentences(sentences)
    sentences = remove_special_characters(sentences)
    sentences = remove_stopwords(sentences)
    x = sentences.split()
    return x
```

```python
#test 5000 sample
samples = full_examples[:5000]
tokenized_corpus = []
for sample in samples:
    sentences = sample.decode('utf-8')
    x = tokenze_sentences(sentences)
    tokenized_corpus.append(x)
    #print(x)
```

```python
#build vocabulary
def build_vocabulary_from_tokenized(tokenized_sentences):
    joined_sentences = [' '.join(sentence) for sentence in tokenized_sentences]
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(joined_sentences)
    vocabulary = vectorizer.get_feature_names_out()
    return vocabulary


def count_words_in_vocabulary(tokenized_sentences, vocabulary):
    joined_sentences = [' '.join(sentence) for sentence in tokenized_sentences]
    vectorizer = CountVectorizer(vocabulary=vocabulary)
    X = vectorizer.fit_transform(joined_sentences)
    word_counts = pd.DataFrame(X.toarray(), columns=vocabulary)
    return word_counts

vocabulary = build_vocabulary_from_tokenized(tokenized_corpus)
word_counts_df = count_words_in_vocabulary(tokenized_corpus, vocabulary)
print(word_counts_df)
```

## Câu 4
Write a program to calculate the term frequency (TF) of each word in the reviews.

```python
def calculate_tf(sentences):
    words = tokenze_sentences(sentences)
    tf_dict = {}
    for word in words:
        tf_dict[word] = tf_dict.get(word, 0) + 1
    total_words = len(words)
    tf_dict = {word: count / total_words for word, count in tf_dict.items()}
    return tf_dict
```

```python
#test
sentence = full_examples[0]
sentence = sentence.decode('utf-8')
tf_result = calculate_tf(sentence)
for word, tf in tf_result.items():
        print(f"{word}: {tf}")
```

## Câu 5
Write a program to calculate the inverse document frequency (IDF) of words in the reviews, considering each review as a separate document.

```python
def calculate_idf(reviews):
    all_words = set()
    for review in reviews:
        sentences = review.decode('utf-8')
        words = tokenze_sentences(sentences)
        all_words.update(words)

    df_dict = {word: 0 for word in all_words}
    # Count the document frequency for each word
    for review in reviews:
        sentences = review.decode('utf-8')
        unique_words_in_review = set(tokenze_sentences(sentences))
        for word in unique_words_in_review:
            df_dict[word] += 1
    # Calculate IDF
    total_reviews = len(reviews)
    idf_dict = {word: math.log(total_reviews / (1 + df)) for word, df in df_dict.items()}
    return idf_dict
```

```python
#test
idf_result = calculate_idf(samples)
for word, idf in idf_result.items():
    print(f"{word}: {idf}")
```

## Câu 6
Write a function that takes a review sentence as input and returns its TF-IDF representation using the information obtained in the previous steps.

```python
def calculate_tf_idf(review, idf_dict):
    words = tokenze_sentences(review)
    tf_review = {word: words.count(word) / len(words) for word in words}
    tf_idf_review = {word: tf * idf_dict.get(word, 0) for word, tf in tf_review.items()}
    return tf_idf_review
```
